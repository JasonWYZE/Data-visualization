```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kde
import seaborn as sns

data = pd.read_csv('HappinessAlcoholConsumption.csv')

data_uniq = pd.unique(data.Region)
print(len(data_uniq))



#Happiness scores box plot
plt.figure(0)
data.HappinessScore.plot(kind='box')
plt.show()

#Happiness score by region
"""
grouped_region = data.HappinessScore.groupby(data.Region)
mean_happinessScore = grouped_region.mean()
region_happinessScore = mean_happinessScore.reset_index()
print(region_happinessScore)
"""
region_happinessScore = data.HappinessScore.groupby(data.Region).mean().reset_index()
region_happinessScore = region_happinessScore.sort_values(by = 'HappinessScore')
x = list(region_happinessScore.HappinessScore)
y = list(region_happinessScore.Region)
plt.figure(1)
plt.barh(y,x)
plt.ylabel('Region')
plt.xlabel('Mean Happiness Score')
plt.title('Happiness Score By Region')
plt.show()

#Top 25 Countries by Happiness Score
country_happinessScore = data.HappinessScore.groupby(data.Country).mean().reset_index()
country_happinessScore = country_happinessScore.sort_values(by = 'HappinessScore', ascending= False, inplace=True).head(25)
print(country_happinessScore)
x = list(country_happinessScore.HappinessScore)
y = list(country_happinessScore.Country)
plt.figure(2)
plt.barh(y,x)
plt.ylabel('Country')
plt.xlabel('Mean Happiness Score')
plt.title('Top 25 Countries by Happiness Score ')
plt.show()

# Human Development Index
plt.figure(3)
sns.kdeplot(data.HDI,color="gray",shade=True)
mean_HDI = np.mean(data.HDI)
plt.axvline(mean_HDI,linestyle='dashed',linewidth=2)
plt.show()

# Human Development Index by Region
region_HDI = data.HDI.groupby(data.Region).mean().reset_index()
region_HDI = region_HDI.sort_values(by = 'HDI', ascending = True)
x = list(region_HDI.HDI)
y = list(region_HDI.Region)
plt.figure(4)
plt.barh(y,x)
plt.ylabel('Region')
plt.xlabel('Mean of HDI')
plt.title('HDI By Region')
plt.show()

#Top 25 Countries by Human Development Index
country_HDI = data.HDI.groupby(data.Country).mean().reset_index()
country_HDI = country_HDI.sort_values(by = 'HDI',ascending=True).head(25)
x = list(country_HDI.HDI)
y = list(country_HDI.Country)
plt.figure(5)
plt.barh(y,x)
plt.ylabel('Country')
plt.xlabel('Mean Human Development Index')
plt.title('Top 25 Countries by Human Development Index')
plt.show()


```
